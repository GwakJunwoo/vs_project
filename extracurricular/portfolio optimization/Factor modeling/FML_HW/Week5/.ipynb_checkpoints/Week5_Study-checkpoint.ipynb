**Portfolio Risk and Return**

$\sigma_{P}^{2}=\sum_{i} p(i)\left[r_{P}-E\left(r_{P}\right)\right]^{2}$

$\begin{aligned} \sigma_{P}^{2} &=\sum_{i} p(i)\left[x_{A} r_{A}-x_{A} E\left(r_{A}\right)+x_{B} r_{B}-x_{B} E\left(r_{B}\right)\right]^{2} \\ &=\sum_{i} p(i)\left[x_{A}\left(r_{A}-E\left(r_{A}\right)\right)+x_{B}\left(r_{B}-E\left(r_{B}\right)\right)\right]^{2} \end{aligned}$

$=x_{A}^{2} \sum_{i} p(i)\left(r_{A}-E\left(r_{A}\right)\right)^{2}+x_{B}^{2} \sum_{i} p(i)\left(r_{B}-E\left(r_{B}\right)\right)^{2}+2 x_{A} x_{B} \sum_{i} p(i)\left(r_{A}-E\left(r_{A}\right)\right)\left(r_{B}-E\left(r_{B}\right)\right)$

$=x_{A}^{2} \sigma_{A}^{2}+x_{B}^{2} \sigma_{B}^{2}+2 x_{A} x_{B} \operatorname{Cov}\left(r_{A}, r_{B}\right)$

$\operatorname{Cov}\left(r_{A}, r_{B}\right)=\mathrm{E}\left[\left(r_{A}-\bar{r}_{A}\right)\left(r_{B}-\bar{r}_{B}\right)\right]$

$=\frac{1}{n-1} \sum_{i=1}^{n}\left(r_{A i}-\bar{r}_{A}\right)\left(r_{B i}-\bar{r}_{B}\right)$

if $\bar{r}_{A}=\bar{r}_{B}=0$, then

$=\frac{1}{n-1} \sum_{i=1}^{n} r_{A i} r_{B i}$

In matrix notation, this equals


$$
\frac{1}{n-1} \mathbf{r}_{A}^{\mathrm{T}} \mathbf{r}_{B}
$$
Therefore, if $\mathbf{r}$ is a matrix that contains the vectors $\mathbf{r}_{A}$ and $\mathbf{r}_{B}$ as its columns,

**Sharpe Ratio**

Sharpe ratio is the ratio of reward to volatility.

Sharpe Ratio $=\frac{r_{\text {risky portfolio }}-r_{\text {risk free }}}{\sigma_{\text {excess return }}}$

the risk premium(which we'll denote with _D

$D_{t}=r_{\text {portfolio, } t}-r_{\text {risk free, } t}$

$$
D_{\text {average }}=\frac{1}{T} \sum_{t=1}^{T} D_{t} \quad \sigma_{D}=\sqrt{\frac{\sum_{t=1}^{T}\left(D_{t}-D_{\text {average}}\right)^{2}}{T-1}}
$$
Sharpe Ratio $=\frac{D_{\text {averas }}}{\sigma_{D}}$ = Capital Market Line

**Annualized Sharpe Ratio**

Sharpe Ratio$_{\text {year }}=\sqrt{252}$ Sharpe Ratio $_{\text {day }}$

$\sigma_{D_{\text {year }}}=\sqrt[2]{252} \times \sigma_{D_{\text {, day }}}$

Sharpe Ratio $_{\text {year }}=\sqrt{252}$ Sharpe Ratio $_{\text {day }}$

**Second-order condition**

in general, points where the derivative equals 0 could be mnima, maxima, or saddle points.

- If $\frac{\mathrm{d}^{2} y}{\mathrm{~d} x^{2}}\left(x_{0}\right)<0$ then $f$ has a local maximum at $x_{0}$.
- If $\frac{\mathrm{d}^{2} y}{\mathrm{~d} x^{2}}\left(x_{0}\right)>0$ then $f$ has a local minimum at $x_{0}$.
- If $\frac{\mathrm{d}^{2} y}{\mathrm{~d} x^{2}}\left(x_{0}\right)=0$, the test is inconclusive.

for a funcion of two variables, the rule must change a bit. We construct the matrix of second=order partial derivatives:

$H(x, y)=\left(\begin{array}{ll}f_{x x}(x, y) & f_{x y}(x, y) \\ f_{y x}(x, y) & f_{y y}(x, y)\end{array}\right)$

this is called the Hessian matrix H(x,y)

$\operatorname{det}\left(\begin{array}{ll}a & b \\ c & d\end{array}\right)=a d-b c$

if the first-order partial derivatives are 0 at the point (a,b), i.e. 
$f_{x}(a, b)=f_{y}(a, b)=0$, then, we apply the following rule:

- If $\operatorname{det}(H)(a, b)>0$ and $f_{x x}(a, b)>0$ then $(a, b)$ is a local minimum of $f$.


- If $\operatorname{det}(H)(a, b)>0$ and $f_{x x}(a, b)<0$ then $(a, b)$ is a local maximum of $f$.


- If $\operatorname{det}(H)(a, b)<0$ then $(a, b)$ is a saddle point of $f$.


- If $\operatorname{det}(H)(a, b)=0$ then the second derivative test is inconclusive, and the point $(a, b)$ could be any of a minimum, maximum or saddle point.

**Formulating Portfolio Optimization Problems**

Common Constraints
1. no short selling: 

$0 \leq x_{i} \leq 1, \quad i=1,2, \ldots, n$

2. sector limits: 

$x_{\text {biotech1 }}+x_{\text {biotech } 2}+x_{\text {biotech } 3} \leq M$, $M$ = percent of portfolio allocations in individual sectors

3. constrain on portfolio return:

$\mathbf{x}^{\mathrm{T}} \mu \geq r_{\min }$, $r_{\min }$ = minimum acceptable portfolio return

4. Maximizing Portfolio Return:

objective: minimize: $-\mathbf{x}^{\mathrm{T}} \mu$

constraint:
$\mathbf{x}^{\mathrm{T}} \mathbf{P x} \leq p, \quad p=$ maximum permissible portfolio variance

5. Maximizing Portfolio Return And Minimizing Portfolio Variance

objective:
minimize : $\quad-\mathbf{x}^{\mathrm{T}} \mu+b \mathbf{x}^{\mathrm{T}} \mathbf{P} \mathbf{x}, \quad b=$ tradeoff parameter.

6. Minimizing Distance to a Set of Target Weights

objective:
minimize: $\quad\left\|\mathbf{x}-\mathbf{x}^{*}\right\|_{2}$, ${x}^{*}$ = a set of target portfolio weights

7. Tracking an Index

objective:
minimize : $\quad \mathbf{x}^{\mathrm{T}} \mathbf{P} \mathbf{x}+\lambda\|\mathbf{x}-\mathbf{q}\|_{2}, \quad \mathbf{q}$ = a set of index weights

```python
# example of use of cvxpy
!pip install cvxpy

import cvxpy as cvx
import numpy as np

x = cvx.Variable(1)
objective = cvx.Minimize((x - 1)**2 + 1)
constraints = [x <= 0]
problem = cvx.Problem(objective, constraints)
result = problem.solve()

print('Optimal value of x: {:.6f}'.format(x.value[0]))
print('Optimal value of the objective: {:.6f}'.format(problem.value))
```
