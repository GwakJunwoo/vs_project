```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import yfinance as yf

class ratio:
    
    def __init__(self,tickers, start):
        data = yf.download(tickers, start=start)
    
    def log_return(self):
        log_price = np.log(self.data.close)
        log_return = (log_price[:-1] - log_price[1:]).to_numpy()
        return pd.Series(log_return, index=log_price[1:].index)
    
    def returns(self):
        prices = self.data.close.to_numpy()
        pre = prices[1:]
        returns = list()
        for i in range(len(pre)):
            gap = pre[i] - prices[i]
            returns.append(gap/prices[i])
        return pd.Series(returns, index=self.data.close[1:].index)
```
